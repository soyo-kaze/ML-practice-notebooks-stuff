### MaxMinScaler

```python
trans_min = sk.MinMaxScaler() # object of maxminscaler is created
```

```python
trans_data = trans_min.fit_transform(data) # fits and transfroms the data at once unlike exclusive fit and transform functions
```

```python
trans_data
```

```python
transd = p.DataFrame(trans_data)
```

```python
transd
```

```python
c = data.columns
```

```python
c = list(c)
```

```python
transd
```

```python
transd.columns = c # changes the columns from defaul to elements defined in the list
```

```python
transd
```

### Robust scaling
* for data outliers

```python
trans_robust = sk.RobustScaler()
```

```python
trans_rd = trans_robust.fit_transform(data)
```

```python
trans_rd = p.DataFrame(trans_rd)
```

```python
trans_rd
```

```python
trans_rd.columns = c
```

```python
trans_rd
```

### Normalizing data by L2 normalization
* For clustering (maybe)

```python
norm = sk.Normalizer() # normalizer object
```

```python
norm_data = norm.fit_transform(data.loc[:1,])
```

```python
p.DataFrame(norm_data)
```

### Data Imputation

```python
from sklearn.impute import SimpleImputer # it is not defined in preprocessing class
```

```python
imp = SimpleImputer(strategy='mean')
```

```python
imp_data = imp.fit_transform(data)
```

```python
imp_data = p.DataFrame(imp_data)
```

```python
imp_data.info() #compare it with data.info() you can see there is no missing data
```
