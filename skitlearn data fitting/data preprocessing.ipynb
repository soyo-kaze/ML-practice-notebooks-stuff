### Data Imputation

```python
from sklearn.impute import SimpleImputer # it is not defined in preprocessing class
```

```python
imp = SimpleImputer(strategy='mean') # strategies = median, most_frequent, (constant, fill_value=x)
```

```python
imp_data = imp.fit_transform(data) # returns numpy array instead of a dataframe
```

```python
imp_data = p.DataFrame(imp_data) # changes that ndarray to a readable dataframe
```

```python
imp_data.info() #compare it with data.info() you can see there is no missing data
```

### Lables

```python
from sklearn.datasets import load_breast_cancer
```

```python
br = load_breast_cancer()
bc = PCA(n_components=2)
```

```python
br_data = p.DataFrame(br.data)
```

```python
brp_data = bc.fit_transform(br_data)
```

```python
brp_names = br.target_names
brp_tar = br.target
```

```python
def get_info(names,b_data,ind,data_ind):
    label_name = names[ind]
    label_data = b_data[data_ind==ind]
    return (label_name,label_data)
```

```python
def seperate_data(b_data,names,data_ind):
    l = []
    for n in range(len(names)):
        l.append(get_info(names,b_data,n,data_ind))
    return l
```

```python
s_da = seperate_data(brp_data,brp_names,brp_tar)
```

```python
for x,y in s_da:
    c1 = y[:,0]
    c2 = y[:,1]
    plt.scatter(c1,c2,label=x)
plt.legend()
```
